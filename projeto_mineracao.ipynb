```python
import pandas as pb
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/ivanjrinacio/Mineracao_Unicesumar/main/dados_brutos.csv'
```

```python
dados_brutos = pb.read_csv(url)
dados_brutos.head(15)
```

```python
df = pb.DataFrame(dados_brutos)
df.info()
df.isnull().sum()
```

```python
df_sem_null = df.dropna()
df_sem_null.info()
df_sem_null.isnull().sum()
```

```python
print("\nVerificando se há duplicatas:")
print(df_sem_null.duplicated(subset='Nome').sum())
df_sem_duplicatas = df_sem_null.drop_duplicates(subset='Nome', keep='first')
print("Forma do DataFrame original (sem nulos):", df_sem_null.shape)
print("Forma do DataFrame sem duplicatas:", df_sem_duplicatas.shape)
print("\nVerificando se ainda há duplicatas:")
print(df_sem_duplicatas.duplicated(subset='Nome').sum())
```

```python
df_limpo = df_sem_duplicatas
df_limpo.info()
df_limpo.head(15)
```

```python
df_limpo.to_csv('dados_limpos.csv', index=False)
```

```python
df_limpo = df_sem_duplicatas.copy()
df_limpo['Evasao'] = df_limpo['Evasao'].map({'Sim': 1, 'Não': 0})
df_limpo['Atividades_Extracurriculares'] = df_limpo['Atividades_Extracurriculares'].map({'Sim': 1, 'Não': 0})
df_limpo['Programa_de_Apoio'] = df_limpo['Programa_de_Apoio'].map({'Sim': 1, 'Não': 0})
df_limpo.head(15)
```

```python
from sklearn.preprocessing import MinMaxScaler

colunas_numericas = ['Nota_Media', 'Frequencia_Porcentagem', 'Absenteismo_Faltas', 'Participacao_Aulas_Porcentagem'] 
scaler = MinMaxScaler()
df_limpo[colunas_numericas] = scaler.fit_transform(df_limpo[colunas_numericas])

df_limpo.head(15)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

colunas_para_converter = ['Nota_Media', 'Frequencia_Porcentagem', 'Absenteismo_Faltas', 'Participacao_Aulas_Porcentagem']

for col in colunas_para_converter:
    df_limpo.loc[:, col] = pb.to_numeric(df_limpo[col], errors='coerce')

colunas_para_agrupar = ['Nota_Media', 'Frequencia_Porcentagem', 'Absenteismo_Faltas', 'Participacao_Aulas_Porcentagem']

X = df_limpo[colunas_para_agrupar]

kmeans = KMeans(n_clusters=3, random_state=42)
df_limpo['cluster_risco'] = kmeans.fit_predict(X)

print(df_limpo.groupby('cluster_risco').mean(numeric_only=True))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = df_limpo.drop(['ID_Aluno', 'Nome', 'Evasao'], axis=1)
y = df_limpo['Evasao']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)

y_pred = tree_model.predict(X_test)
print(f"Acurácia do modelo: {accuracy_score(y_test, y_pred):.2f}")
```

```python
X = df_limpo.drop(['ID_Aluno', 'Nome', 'Evasao', 'cluster_risco'], axis=1)
y = df_limpo['Evasao']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)

probabilidades_evasao = tree_model.predict_proba(X)[:, 1]

df_limpo['Score_Risco_Evasao'] = probabilidades_evasao * 100

print("Alunos com maior risco de evasão:")
print(df_limpo.sort_values(by='Score_Risco_Evasao', ascending=False).head(10))
```

```python
df_analise = df_limpo.copy()

df_analise = df_analise.drop(['ID_Aluno', 'Nome', 'Score_Risco_Evasao'], axis=1)

cores_clusters = ['#1f77b4', '#ff7f0e', '#2ca02c']

cluster_means = df_analise.groupby('cluster_risco').mean(numeric_only=True)

cluster_means[['Nota_Media', 'Absenteismo_Faltas']].T.plot(kind='bar', figsize=(10, 6), color=cores_clusters)
plt.title('Média da Nota e Faltas por Cluster de Risco')
plt.xlabel('Atributos')
plt.ylabel('Valores Médios')
plt.xticks(rotation=0)
plt.legend(title='Cluster')
plt.show()

porcentagem_cols = ['Frequencia_Porcentagem', 'Participacao_Aulas_Porcentagem', 
                    'Atividades_Extracurriculares', 'Programa_de_Apoio', 'Evasao'] 

plt.figure(figsize=(15, 10)) 

for i, col in enumerate(porcentagem_cols):
    plt.subplot(2, 3, i + 1) 
    cluster_means[col].plot(kind='bar', rot=0, color=cores_clusters)
    plt.title(f'Média de {col} por Cluster')
    plt.ylabel('Média')
    plt.xlabel('Cluster de Risco')
    plt.ylim(0, 1) 

plt.tight_layout() 
plt.show()
```

```python
# Mapear os rótulos de cluster numéricos para nomes descritivos
cluster_nomes = {0: 'Baixo Risco', 1: 'Alto Risco', 2: 'Médio Risco'}
# Crie uma nova coluna com os nomes dos clusters
df_limpo['cluster_nome'] = df_limpo['cluster_risco'].map(cluster_nomes)

plt.figure(figsize=(10, 6))
# Plote o gráfico usando a nova coluna de nomes
scatter = plt.scatter(df_limpo['Absenteismo_Faltas'], df_limpo['Nota_Media'], c=df_limpo['cluster_risco'], cmap='viridis')
plt.title('Relação entre Faltas e Notas, por Cluster de Risco')
plt.xlabel('Absenteísmo (Faltas)')
plt.ylabel('Nota Média')
plt.legend(*scatter.legend_elements(), title='Cluster de Risco')
plt.show()
```
